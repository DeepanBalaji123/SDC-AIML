```python
from transformers import pipeline

# Load the summarization pipeline with the BART model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Example long text
text = """
Climate change is having a significant impact on global weather patterns. Rising temperatures have led to more frequent
and severe heatwaves, droughts, and hurricanes. Glaciers are melting at unprecedented rates, contributing to rising sea levels
that threaten coastal communities. These changes are affecting agriculture, biodiversity, and water supplies, posing serious
challenges for governments and citizens worldwide.
"""

# Generate summary using the summarizer pipeline
summary = summarizer(text, max_length=100, min_length=50, do_sample=False)

# Print the summary
print("📝 Summary:")
print(summary[0]['summary_text'])
```
